## Оценка надежности с применением формулы Спирмена-Брауна

```python
data.columns = range(14)
data.rename(columns={0: 'education', 13:'total score'}, inplace=True)
data
```

Напомним, что для расчета корреляции используется следующих коэффициент:

$$r_{1/2} = \frac{n  \sum_{i=1}^n x_i y_i - \left( \sum_{i=1}^n x_i\right) \cdot \left( \sum_{i=1}^n y_i\right) }{\sqrt{\left( n\sum_{i=1}^n x_i^2 - \left(\sum_{i=1}^n x_i\right)^2 \right) \cdot \left( n\sum_{i=1}^n y_i^2 - \left(\sum_{i=1}^n y_i\right)^2 \right)}}$$

где $x_i$ и $y_i$ - баллы испуытуемого $i$ за нечетные и четные задания соответственно. То, что в тесте были задания с единственным верным вариантом ответа, которые мы не учитываем, по сути, не должно влиять на результат, так как задания были разбросаны в случайном порядке, что не должно было как-то особоенно повлиять на итоговый балл испытуемых. 

Для внесения большей ясности введем обозначим отдельные компоненты указанной выше формулы:
$$ r_{1/2} = \frac{A - B \cdot C }{\sqrt{\left( D - B^2 \right) \cdot \left( E - C^2 \right)}}$$

Рассчитаем компоненты:

```python
scores = np.array(data[range(1, 13)])
X = np.sum(scores[:, ::2], axis=1)
Y = np.sum(scores[:, 1::2], axis=1)
n = len(X)
print(X)
print(Y)
```

```python
A = n * X @ Y.T
B = np.sum(X)
C = np.sum(Y)
D = n * np.sum(X ** 2)
E = n * np.sum(Y ** 2)

r0 = (A - B * C) / np.sqrt((D - B ** 2) * (E - C ** 2))
r0
```

Теперь сделаем корректировку показателя с помощью формулы Спирмена-Брауна:
$$ r_t = \frac{2r_{1/2}}{1 + r_{1/2}} $$

```python
rt = 2 * r0 / (1 + r0)
rt
```

Полученный коэффициент является весьма высоким, что вызвано, скорее, выборкой, нежели качеством заданий. Несмотря на то, что показатель превышает пороговый (0,7), точный вывод о надежности составленной совокупности заданий для множественного выбора сделать нельзя. В представленнной для анализа выборке было мало наблюдений, cлишком много ответов участников были неправильными, из-за чего коэффициент корреляции получился довольно высоким (так как ноль коррелирует с нулем идеально).

Тем не менее, здесь представлен достаточно подробный и, надеюсь, понятный способ получения оценки надежности теста, который можно импользовать в более крупных исследованиях. А приведенные расчеты можно использовать в качестве ценного примера и напоминания, что получение благоприятных вычисленных показателей не всегда однозначно соответствует благоприятный результат.

### Для тестов с едининственным верным варианта ответа

В исследовательских целях рассчитаем теперь коэффициент надежности для заданий с одним верным вариантом ответа.

```python
multiple_numbers = [3, 4, 6, 7, 10, 11, 13, 15, 19, 20, 22, 23]
d1 = df1.drop(multiple_numbers, axis='columns')
d1.head()
```

```python
d1.columns = range(16)
d1.rename(columns={0: 'education', 15:'total score'}, inplace=True)
d1.head()
```

```python
scores1 = np.array(d1[range(1, 15)])
X1 = np.sum(scores1[:, ::2], axis=1)
Y1 = np.sum(scores1[:, 1::2], axis=1)
n1 = len(X1)
print(X1)
print(Y1)
```

```python
A = n1 * X1 @ Y1.T
B = np.sum(X1)
C = np.sum(Y1)
D = n1 * np.sum(X1 ** 2)
E = n1 * np.sum(Y1 ** 2)

r1= (A - B * C) / np.sqrt((D - B ** 2) * (E - C ** 2))
rt1 = 2 * r1 / (1 + r1)
rt1
```

Получилось, чо надежность тестов с одним верным вариантом ответа меньше. Но опять же, скорее всего, это связано с тем, что участникам тестирования было проще отвечать на такие вопросы, количество полученных нулей за эти задания меньше, по сравнению с вопросами на множественный выбор ответа

### Корреляция баллов участников за задания с одиночным и множественным выбором верного варианта ответа

```python
X2 = X1 + Y1
Y2 = X + Y
n2 = len(X2)
Y2, X2
```

```python
A = n2 * X2 @ Y2.T
B = np.sum(X2)
C = np.sum(Y2)
D = n2 * np.sum(X2 ** 2)
E = n2 * np.sum(Y2 ** 2)

r2 = (A - B * C) / np.sqrt((D - B ** 2) * (E - C ** 2))
r2
```

```python
2 * r2 / (1 + r2)
```

Коэффициент корреляции вновь получился высоким, что и должно было получаться из логики: если участник обладает достаточными знаниями, то он будет хорошо отвечать на задания обоих типов, и наоборот: если участник имеет плохие знания, то вряд ли его полученные низкие баллы за задания двух типов будут различаться. Но опять же стоит отметить, что в выборке многие участники получали нули, которые, вероятно, и влияют на столь большую величину показателя
